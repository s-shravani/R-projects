```r
vec_marital <- table(data$marital)
```

```r
vec_marital
```

```r
percent_marital <- round(100*vec_marital/sum(vec_marital),1)
```

```r
percent_marital
```

```r
pie_labels <- paste(percent_marital,"%",sep = "")
```

```r
pie(percent_marital,col=rainbow(length(vec_marital)), labels = pie_labels,main = "Marital Status Percentage",cex=0.8)
legend("topright",c("Divorced","Married","Single"),cex = 0.7, fill =rainbow(length(vec_marital)))
```
